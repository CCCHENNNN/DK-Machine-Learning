```python
import pandas as pd
import numpy as np

df = pd.read_csv('all/train-set.csv')
```

```python
df.head(10)
```

```python
df['Elevation']=df['Elevation'].map(lambda x: 'x' if x<3100 and x>2900 else '7' if x<3200 and x>3100 else '6')
df = pd.get_dummies(df)
df.info()
```

```python
train_data = df.values
```

```python
print(train_data)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 10)
```

```python
model = model.fit(train_data[0:,1:-1],train_data[0:,-1])
```

```python
df_test = pd.read_csv('all/test-set.csv')
```

```python
df_test.head(10)
```

```python
test_data = df_test.values
```

```python
output = model.predict(test_data[:,1:])
```

```python
print(output)
```

```python
from sklearn.metrics import accuracy_score
```

```python
result = np.c_[test_data[:,0].astype(int), output.astype(int)]
df_result = pd.DataFrame(result[:,0:2], columns=['Id', 'Cover_Type'])
```

```python
print(df_result)
```

```python
df_result.to_csv('all/forest.csv', index=False)
```
